```python
def change_time(x):
  x = str(x).split(".")[0]
  if(len(x) == 3):
    return str(x[0] + ":" + x[1:] + ":00")
  elif(len(x) == 4):
    if x[0:2] == "24":
      x = "00" + x[2:]
    return str(x[0:2] + ":" + x[2:] + ":00")
  else:
    return str("00:" +x + ":00")

df_final_copy["WHEELS_OFF"] = df_final_copy["WHEELS_OFF"]. apply(change_time)
df_final_copy["WHEELS_ON"] = df_final_copy["WHEELS_ON"]. apply(change_time)
gc.collect()
df_final_copy["ARR_TIME"] = df_final_copy["ARR_TIME"]. apply(change_time)
gc.collect()
df_final_copy["DEP_TIME"] = df_final_copy["DEP_TIME"]. apply(change_time)
```
